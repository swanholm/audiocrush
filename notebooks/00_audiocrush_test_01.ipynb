```python
import torch
import sys
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display

from audiocrush.core import TensorAudio
```

```python
def generate_sine_wave(freq, duration, sample_rate=32000):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    waveform = np.sin(2 * np.pi * freq * t)
    return waveform
```

```python
freq = 440  # Frequency in Hz
duration = 2.0  # Duration in seconds
sample_rate = 32000  # Samples per second

waveform = generate_sine_wave(freq, duration, sample_rate)
audio_tensor = TensorAudio(torch.tensor(waveform, dtype=torch.float32), sample_rate=sample_rate)
```

```python
freq2 = 880  # Frequency in Hz

waveform2 = generate_sine_wave(freq2, duration, sample_rate)
audio_tensor2 = TensorAudio(torch.tensor(waveform2, dtype=torch.float32), sample_rate=sample_rate)
```

```python
combined_waveform = waveform + waveform2
audio_tensor3 = TensorAudio(torch.tensor(combined_waveform, dtype=torch.float32), sample_rate=sample_rate)
```
